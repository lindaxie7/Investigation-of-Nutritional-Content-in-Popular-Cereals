```python
# Merge ingredient and cerealingredient
merge_df_1 = pd.merge(df_ingredient, df_cerealingredient, on="ingredientid", how="left")
merge_df_1
```

```python
# Merge creal, ingredient, cerealingredient
merge_df_2 = pd.merge(merge_df_1 , df_cereal, on="upc", how="left")
merge_df_2
```

```python
# Rename columns so that they are differentiated

merge_df_2 = merge_df_2.rename(columns={ "name_x": "ingredientname",  "name_y": "cerealtname" })

merge_df_2.head()
```

```python
# rename id to nutrientid in nutrient, so it can be merged with cerealnutrient
df_nutrient = df_nutrient.rename(columns={"id": "nutrientid", "name": "nutrientname"})
df_nutrient
```

```python
# Merge nutrient, crealnutrient
merge_df_3 = pd.merge(df_nutrient, df_cerealnutrient, on="nutrientid", how="left")
merge_df_3
```

```python
# Merge nutrient, crealnutrient, creal, ingredient, cerealingredient
merge_df_4 = pd.merge(merge_df_2, merge_df_3, on="upc", how="left")
merge_df_4
```

```python
# rename id to manufacturerid in manufacturer , so it can be merged with merge_df_4
df_manufacturer = df_manufacturer.rename(columns={"id": "manufacturerid", "name": "manufacturername"})
df_manufacturer
```

```python
# Merge manufacturer, nutrient, crealnutrient, creal, ingredient, cerealingredient
merge_df_5 = pd.merge(df_manufacturer, merge_df_4, on="manufacturerid", how="left")
merge_df_5
```

```python
# rename amount
merge_df_5 = merge_df_5.rename(columns={"amount": "nutrientamount"})
merge_df_5
```

```python
# rename columns in allergy
df_allergy = df_allergy.rename(columns={"name": "allergyname", "type": "allergytype", "origin": "foodorigin"})
df_allergy
```

```python
# Merge allergy and merge_df_4
merge_df_6 = pd.merge(df_allergy, merge_df_5, on="ingredientid", how="outer")
merge_df_6
```
